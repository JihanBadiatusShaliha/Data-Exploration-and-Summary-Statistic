```python
 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Anggap df adalah DataFrame yang sudah kamu definisikan

# Hanya pilih kolom numerik
numerical_df = df.select_dtypes(include=['number'])

# Loop untuk visualisasi histogram, box plot, dan scatter plot
for column in numerical_df.columns:
    plt.figure(figsize=(15, 6))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(numerical_df[column], kde=True, color='pink')
    plt.title(f'Histogram of {column}')

    # Box plot
    plt.subplot(1, 2, 2)
    sns.boxplot(y=numerical_df[column], color='pink')
    plt.title(f'Box Plot of {column}')

    plt.show()

# Scatter plot antara pasangan variabel numerik
sns.pairplot(numerical_df)
plt.show()

# Heatmap korelasi antar variabel numerik
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap Korelasi Antar Variabel Numerik')
plt.show()
```
